### Ways to access files

The basic way to open files is to use `open(file, access_mode)`.

Whenever we use `open` we accompany it with an access mode, which sets the pattern for how the `file` will be used once it is loaded into memory. 

* Read only
    * Access mode: `r`
* Read and write
    Access mode: `r+` * Access mode: `r+`
* Write only 
    * Access form: `w` * Write only
* Write and read
    * Access form: `w+` * Append only
* Append only
    * Append information (at the end of what exists). 
    * Access form: `a` * Append and read
* Append and read
    * Append information (at the end of whatever exists) in addition to read. 
    Access form: `a+` * Append and read

```python
# Set our directory
path = '.'
file_name='text_file.txt'

# Let's join both variables together to make a single path
basedir= os.path.join(path, file_name)
print(basedir)

# Reading the file using "open"
my_file = open(basedir,'r')

# The read command reads everything contained in the file "my file".
print(my_file.read())

# Frees memory
my_file.close()
```

##### Using `readline`.
Reads specific number of bytes from the line. Does not read more than one line even if the bytes exceed.

```python
my_file = open(basedir,'r')
print(my_file.readline(5))
my_file.close()
# The 5 indicates the number of total bytes to read.
```

## Comma Separated Values (CSV)

We will use `with` from now on. `with` simplifies the resource handling by making the code clearer. When using this tool we do not need to use `.close()`. 

It is used together with `open`, simplifies exceptions and automatically closes the file. 

CSV reading:
https://www.geeksforgeeks.org/reading-and-writing-csv-files-in-python/?ref=rp

Using `with`:
https://www.pythonforbeginners.com/files/with-statement-in-python


Reading with `csv` package.


```python
basedir = os.path.join(path,'csv_file.csv')

with open(basedir,'r') as f:
    my_file = csv.reader(f) #can be read line by line. 
    
    # reading line by line
    for line in my_file:
        print(line)
```

```python
# Mandar esto a volar y usemos pandas
df =pd.read_csv(basedir)
df
```

```python
# Write in the csv that we just read
df['my_column']= 7
df

# Save it
#df.to_csv(#Path)
df.to_csv(basedir)

# Re-read my file to see that my new column has been written
df = pd.read_csv(basedir)
df
```

##### Writing a json

```python
example = {'key': 'value'}
```

```python
with open('example.json','w')as f:
    json.dump(example, f)
```

```python
# We read back
with open('example.json','r')as f:
    new_json = json.load(f)
```
